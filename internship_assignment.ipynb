7. Create Key Metrics

````python
# Daily PnL per account
daily_pnl = (merged_df.groupby(['Account', 'date'])['Closed PnL'].sum().reset_index(name =  'daily_pnl'))

# win rate
merged_df['win'] = merged_df['Closed PnL'] > 0
win_rate = (merged_df.groupby(['Account', 'date'])['win'].mean().reset_index(name = 'win_rate'))

# Average trade size 
# avg_trade = (merged_df.groupby(['Account', 'date']).size().reset_index(name = 'avg_trade'))
avg_trade = (merged_df.groupby(['Account', 'date'])['Size USD'].mean().reset_index(name='avg_trade_size'))

# number of trades per day
trades_per_day = (merged_df.groupby(['Account', 'date']).size().reset_index(name='trades_per_day'))

# Long/Short ratio per day
long_short = (merged_df.groupby(['Account', 'date', 'Side']).size().unstack(fill_value=0).reset_index())

if 'BUY' in long_short.columns and 'SELL' in long_short.columns:
    long_short['long_short_ratio'] = (
        long_short['BUY'] / (long_short['SELL'] + 1e-9)
    )
else:
    long_short['long_short_ratio'] = None
````

````python
print("Daily PnL:\n", daily_pnl)
print("Win Rate:\n", win_rate)
print("Trades per day:\n", trades_per_day)
````

The second line calculates the win rate by taking the mean of values in the 'win' column and considering True -> 1 and False -> 0. 
win rate can be used to analyze whether or not traders are more successful during a certain sentiment.

```python

```
Average trade takes into consideration the risk taking behavior of the trader, if they are willing to take more risks and make larger trades during certain days. 

```python

```
This line counts the number of trades an account made on each day. It can be used to know if traders are more active in terms of trading on certain days ('Fear' days or'Greed' days).

This line counts the number of trades based on 'BUY' or 'SELL'. The values of 'Side' are then converted into columns. It calculates a ratio that represents if traders become more long or short biased during 'Fear' days or 'Greed' days. 

In all these metrics, data was aggregated around 'Account' and 'date' as there exist multiple rows for the same trader at different times of the day and hence there is a presence of too many rows. Aggregating the data leads to crunching down the meticulous details into a concise value that is easy to undertand and still retains the information of the data that is aggregated. Here aggregation is used to convert the multiple rows of a trader or certain types of trade into a single row. 

````python
metrics = daily_pnl.merge(win_rate, on=['Account', 'date'], how='left')
metrics = metrics.merge(avg_trade, on=['Account', 'date'], how='left')
metrics = metrics.merge(trades_per_day, on=['Account', 'date'], how='left')

print("\nSample of final metrics:")
print(metrics.head())
````

This analysis measures the directional trading behavior of traders under different market sentiment conditions. Trades are grouped by sentiment category (Fear, Greed, etc.) and by trade side (BUY or SELL). The number of long (BUY) and short (SELL) trades is counted for each sentiment.

A long/short ratio is then calculated by dividing the number of BUY trades by the number of SELL trades. This ratio indicates whether traders are more bullish or bearish during each sentiment:

A ratio greater than 1 indicates a long (bullish) bias.

A ratio close to 1 indicates balanced positioning.

A ratio less than 1 indicates a short (bearish) bias.

The results are also visualized using a bar chart that compares the number of long and short trades across sentiment categories, making it easier to observe changes in directional behavior.

<h2>Strategy ideas</h2>

````python
#performance by sentiment

if "win" not in merged_df.columns:
    merged_df['win'] = merged_df['Closed PnL']>0

perf = (merged_df.groupby("classification")
        .agg(
            avg_pnl = ("Closed PnL", "mean"),
            win_rate = ("win", "mean"),
            avg_trade_size = ("Size USD", "mean"),
            trades = ("Closed PnL", "count")
        )
        .reset_index()
        )

print("Performance by sentiment:")
print(perf)
````

````python
fear = perf[perf["classification"] == "Fear"]
greed = perf[perf["classification"] == "Greed"]

rules = []

if not fear.empty and not greed.empty:
    fear_pnl = fear["avg_pnl"].values[0]
    greed_pnl = greed["avg_pnl"].values[0]

    fear_size = fear["avg_trade_size"].values[0]
    greed_size = greed["avg_trade_size"].values[0]

    if fear_pnl < greed_pnl:
        rules.append("During Fear days, reduce position sizes or trade less frequently.")
    else:
        rules.append("During Fear days, opportunities appear stronger  consider active trading.")

    if greed_size > fear_size:
        rules.append("During Greed days, traders increase size  use stricter risk controls.")
    else:
        rules.append("During Greed days, no size increase  maintain normal strategy.")

print("Suggested Strategy Rules:")
for i, rule in enumerate(rules, 1):
    print(f"{i}. {rule}")
````

<h1>Methodology</h1>
The objective of this assignment is to study how the market sentiment affects the trader performance and behaviour on Hyperliquid. Two datasets were used namely: fear_greed_index.csv and historical_data.csv. The assignment included three stages: Data Preparation, Metric Construction and Comparative Analysis.

<h2>1. Data Preparation</h2>
<h3>Data preprocessing</h3>
First both the datasets were loaded using pandas. Next, missing values were checked for all columns and removed using forward fill. Forward fill is basically fills in the missing values by substituting the value that comes before the missing value. This is done as forward fill assumes that the state of the data remains unchanged if a missing value is present. Then the duplicate rows are checked for. In case any duplicate rows are present, the are dropped from the dataset. Next, the timestamp column is converted to datetime. This is done in order to make the data easier to understand as well as perform aggregate functions effectively. A new column called 'date' is added to both the datasets after the timestamp conversion. Then both the datasets are merged along the column 'date'. This is done in order to relate trading behaviour and performance to market sentiment. 

<h3>Data Aggregation and metrics</h3>
The combined dataset contained the individual trade records. These records needed to be combined into a higher level of representation of data as the analysis can only be done if some higher level metrics are present. 
The following metrics were calculated:
* 'daily_pnl' represents a daily record of profit and loss of the trader. It can be used to analyze whether traders make more money on Fear or Greed days (Fear and Greed are the sentiments of the traders on particular days).
* 'win' calculates the win rate can be used to analyze whether or not traders are more successful during a certain sentiment.
* 'avg_trade' trade takes into consideration the risk taking behavior of the trader, if they are willing to take more risks and make larger trades during certain days.
* 'trades_per_day' counts the number of trades an account made on each day.
* 'long_short' counts the number of trades based on 'BUY' or 'SELL' and calculates their ratio on a per day basis.

After this, all the metrics are combined with the original dataset and then saved as a new .csv file along with a separate .csv file for daily metrics.

<h2>2. Analysis</h2>
The second section focuses on answering a few questions about the aggregated data. First a performance dataframe is created that includes parameters like, avg_pnl that takes the average of the closed PnL values, 'total_pnl' calculates the total PnL across a certain sentiment by taking the sum of the closed PnL column, 'win_rate' calculates the number of traders that succeed in the trade and is calculated by taking the mean of the win column and, 'trade_count' which calculates the number of trades in each sentiment. 
This enables us to make a direct comparison of both the performance and the behavior across different market conditions.
A few graphs are plotted like Average PnL vs Sentiment, Total PnL vs Sentiment, Win Rate vs Sentiment, Average Trade Size vs Sentiment. This is done in order to get a visula representation of the data and metrics that we calculated. 
Traders were also divided according to their trading styles like frequent vs infrequent traders that is based on the median number of trades per account (traders above median were labelled as frequent), consistent vs inconsistent winners that is based on the overall win rate per account, high vs low levarage traders that is based on average trade size (in USD) where traders above the median trade size were labelled as high leverage. Performance metrics were then considered in the context of these segments.

<h2>3. Actionable Output</h2>
The third section's aim was to propose strategy ideas based on findings. First, the performance metrics like PnL, win rate, and average trade size were first aggregated by grouping along sentiment and comparing across the same. 
Based on the comparisons, 
Sentiment conditions with higher average PnL and win rates were interpreted as more favourable trading environments. 
Conditions with lower win rates or higher activity were interpreted as riskier environments.
Simple rule-based strategy ideas were then derived from these observations. The rules focused on:

Adjusting position size

Changing trade frequency

Applying stricter risk controls under certain sentiment conditions

<h3>Optional- Predictive Modelling</h3>
A predictive model was developed to predict whether a trader would have a profitable day or not. The model uses win rate, average trade size, trade frequency, and sentiment classification. These variables were trained based on ensemble leanring models like RandomForestClassifier where they would predict if the PnL for the next day would be positive or negative. 

<h3>Optional- Clustering of traders into behavioral archetypes</h3>
Considering the features like average PnL, win rate, average trade size, and trade frequency, the traders were divided into clusters using a clustering algorithm called KMeans. Further the traders were clustered into archtypes like high-frequency vs low-margin traders, low-frequency vs high-margin traders and consistent performers.

<h3>Strategy Recommendations</h3>

1. Increase exposure during Greed conditions
According to the analysis, we can infer that the average profit per trade and the win rate is higher for Greed and Fear sentiments. Hence we can increase the position sizes during Greed and Extreme Greed periods. Higher values of average PnL and win rates are an indication of better trade quality during optimistic sentiment. Increasing exposure allows traders to trade with respect to stronger trends. 

2. Reduce position sizes during Fear
Fear has the largest number of trade counts and the highest total PnL value but despite this the win rate is only 41% which is lower than Greed and Extreme Greed. Higher trade counts but lower win rates are an indication of unreliability in trade. Hence we can reduce position sizes during Fear conditions and only focus on high confidence setups.

3. Avoid neutral sentiment for meticulous strategies
Neutral sentiment has the lowest average PnL ($22.22) and lowest win rate (31%). Hence meticulous strategies should be avoided as due to low win rate, it will lead to wasteage of resources due to a lack of direction in the market. 

4. Consistency over high-frequnecy or high-leverage approaches
Consistent traders have higher avergae PnL and win rates as compared to high-frequency or high-leverage traders. Hence the focus should be on consistent and high quality trades than increasing trade frequnecy. Stable strategies outperform high-frequency or high-leverage approaches.
